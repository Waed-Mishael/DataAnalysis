```python
merged_custumer_reciept_df = pd.merge(df_custumer, df_Sales_reciepts, on='customer_id', how='inner')
# Calculate the total number of unique customers
total_customers = merged_custumer_reciept_df['customer_id'].nunique()
print(total_customers)
```

```python
# Filter the DataFrame to include only customers who have made more than one visit
repeat_customers_df = merged_custumer_reciept_df['customer_id'].value_counts().loc[lambda x: x > 1]

# Calculate the number of visits per repeat customer
repeat_visits = repeat_customers_df - 1  # Subtract 1 because the first visit is not counted as a return

# Calculate the average number of visits
average_return_visits = repeat_visits.mean()

print("Average number of customers returning two times:", average_return_visits)
```

```python
# Initialize lists to store visit counts and corresponding percentages
visit_counts = []
percentages = []

# Loop through different visit counts from 1 to 100
for visit_count in range(1, 31):
    # Filter the DataFrame to include only customers who have made more than or equal to visit_count visits
    customers_returning_count = (merged_custumer_reciept_df['customer_id'].value_counts() == visit_count).sum()

    # Calculate the percentage of customers returning equal to or more than visit_count times
    percentage_returning_count = (customers_returning_count / total_customers) * 100

    # Append visit count and percentage to lists
    visit_counts.append(visit_count)
    percentages.append(percentage_returning_count)

# Plot the line graph
plt.plot(visit_counts, percentages)
plt.xlabel('Visit Count')
plt.ylabel('Percentage of Customers')
plt.title('Percentage of Customers Returning in one month')
plt.grid(True)
plt.show()
```

```python

# Assuming 'merged_df' contains the merged DataFrame with customer visits data

# Sort the DataFrame by customer_id and transaction_date
merged_df_sorted = merged_custumer_reciept_df.sort_values(by=['customer_id', 'transaction_date'])

# Initialize an empty list to store time differences
time_diffs = []

# Group the DataFrame by customer_id
grouped = merged_df_sorted.groupby('customer_id')

# Iterate over each group (customer)
for customer_id, group in grouped:
    # Filter the group to include only return visits (visits after the first visit)
    return_visits = group.iloc[1:]

    # Calculate the time difference between consecutive return visits
    time_diff = return_visits['transaction_date'].diff().dt.total_seconds().mean()

    # Append the time difference to the list
    time_diffs.append(time_diff)

# Calculate the average time difference between return visits across all customers
average_time_diff = pd.Series(time_diffs).mean()

# Convert the average time difference from seconds to days for better readability
average_time_diff_days = average_time_diff / (60 * 60 * 24)
print("Average time between return visits (in days):", average_time_diff_days)
```

```python
import pandas as pd

# Convert 'transaction_date' to datetime
df_merged_with_generation['transaction_date'] = pd.to_datetime(df_merged_with_generation['transaction_date'])

# Sort DataFrame by 'customer_id' and 'transaction_date' to ensure proper calculation
df_merged_with_generation = df_merged_with_generation.sort_values(by=['customer_id', 'transaction_date'])

# Calculate time difference between consecutive visits for each customer
df_merged_with_generation['time_diff'] = df_merged_with_generation.groupby('customer_id')['transaction_date'].diff().dt.days

# Drop NaN values (first visit of each customer)
df_merged_with_generation.dropna(subset=['time_diff'], inplace=True)

# Assuming 'generation' is a column in the DataFrame representing the generation of each customer

# Group by 'generation' and calculate the average time difference
average_time_between_visits_by_generation = df_merged_with_generation.groupby('generation')['time_diff'].mean()

print(average_time_between_visits_by_generation)
```
